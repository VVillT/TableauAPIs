## Data Frame Manupilation:
    1. Clean Column names
    2. Add in

    3. Remove []
    4. Add in datasource name

```python
ds2 = pd.DataFrame([f for f in wb.datasources[0].fields.keys()],columns = ['name'])
ds = ds2.join(ds1)
ds.columns
```

```python
#Clean Column Names
ds.columns = [col.replace('_','') for col in ds.columns]
ds.columns
```

```python
ds['name'] = ds['name'].str.replace('[','').str.replace(']','')
ds['name'].sample(7)
```

```python
# Add in data sources and files
#print(ds.caption)    #print(ds.connections)
#print(ds.name)    #print(ds.version)
ds['datasoure_caption'] = str(wb.datasources[0].caption)
ds['datasoure_connections'] = str(wb.datasources[0].connections)
ds['datasoure_name'] = str(wb.datasources[0].name)
ds['datasoure_version'] = str(wb.datasources[0].version)
ds['from_wb'] = str(wb.filename)
ds.sample(4).transpose()
```
